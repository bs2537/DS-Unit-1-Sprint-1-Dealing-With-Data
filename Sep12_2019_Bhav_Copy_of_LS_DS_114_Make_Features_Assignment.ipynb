```python
# clean emp_title
# look at top 20 titles

df['emp_title'].value_counts(dropna=False).head(20)
```

```python
#clean the title and handle missing values
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return'unknown'
  
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

********Working with dates********Very Important part of data engineering

```python
# next step is working with dates

df['issue_d']
```

```python

# extracting issue date in numerical format from issue_d column**** very important step for working with dates******
df['issue_id'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
```

```python
df['issue_id']
# the output is as intended
```

```python
# extracting week day in numerical format from issue_id column

df['issue_id'].dt.weekday
```

```python
# extracting month in numerical format from issue_d column

df['issue_id'].dt.month
```

```python
# creating a new column with issue month in numerical format from above output

df['issue_month'] = df['issue_id'].dt.month
```

```python
df.head()
# the output is as intended with a new issue month column created in numerical format
```

```python
# # creating a new column with issue year in numerical format from above output
df['issue_year'] = df['issue_id'].dt.year
df.head()
# the output is as intended and I have created a new column with year in numerical format
```

```python
# changing the display format of the date to U.S. format (in numerical form)

df ['issue_date_U.S.'] = df['issue_id'].dt.strftime('%m %d %Y') 
df.head()

# the output is as intended
```

```python
import re
df['term_integar'] = df['term'].apply(lambda x: re.search(r'\d+', x).group())
df.head(500)

# this worked and a new column was created with only integar values created from the term column.
```

```python
df.term_integar
```

```python
df.loan_status
```

```python

loan_status_is_great = {"loan_status": {"Current": 1, "Fully Paid": 1, "Late (31-120 days)": 0, "Late (16-120 days)": 0, "Charged Off": 0}}

df.replace(loan_status_is_great, inplace=True)
df.head()
                        
# The loan status column in the pandas dataframe is now changed to numerical values in the output given below.                       
```

```python
df.loan_status
```

```python
# Thus, the output  is as intended (see above) and the loan-status has been changed to the binary form which can be used for further analysis.
```
